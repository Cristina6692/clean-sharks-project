```python
#borrar las dos ultimas columnas cuyos datos no me interesan 
to_drop=['Unnamed: 22','Unnamed: 23']
gsai1 = gsai.drop(to_drop, axis=1)
gsai1.columns
#a partir de ahora trabajaremos sobre gsai1
```

```python
#ordenamos por 'Country'
gsai1= gsai1.sort_values('original order')

```

```python
#Hacemos un nuevo DF (gsai2) solo con las columnas que pueden interesarnos para la hipotesis
gsai2 = gsai1[['Case Number', 'Date', 'Year', 'Type', 'Country', 'Area', 'Name', 'Sex ', 'Injury', 'Fatal (Y/N)', 'original order']]
#Y unificamos los nombres de las columnas
gsai2 = gsai2.rename(columns={'Sex ':'Sex', 'Fatal (Y/N)':'Fatal','original order':'Original Order' })
gsai2.head()
```

##COLUMNA FATAL##

```python
gsai2['Fatal'].isnull().sum()
```

```python
gsai2['Fatal'] = gsai2['Fatal'].fillna('UNKNOWN')
```

```python
#Miramos los diferentes valores que tiene la columna original 'Fatal (Y/N)'
#Queremos acotar todos los valores a 'N', 'Y' o 'UNKNOWN'
gsai1['Fatal (Y/N)'].value_counts()
```

```python
#reemplazamos los antiguos valores por los que corresponderian 
gsai2['Fatal'] = gsai2['Fatal'].replace({'N ': 'N', ' N': 'N', 'n':'N', '#VALUE!': 'UNKNOWN', 'F':'Y'}) 
```

```python
#ahora tenemos solo los tres valores que queriamos para nuestra columna
gsai2['Fatal'].value_counts()
```

##COLUMNA SEX##

```python
#vemos cuantos valores son nulos
gsai2['Sex'].isnull().sum()
```

```python
#no queremos que los valores sean nulos por lo que os sustituimos por '-'
gsai2['Sex'] = gsai2['Sex'].fillna('-')
```

```python
gsai2['Sex'].isnull().sum()
```

```python
#localizamoso los valores '-'
gsai2.loc[gsai2['Sex']=='-'].head(10)
```

```python
gsai1['Sex '].value_counts()
```

```python
#miramos los vaores de 'Sex' y unificamos 
keysSex = dict(gsai2['Sex'].value_counts()).keys()
```

```python
def cambioSexKey(x):
    for key in x:
        if key != 'M' and key != 'F' and key!='-':
            if key == 'M ' or key == 'N':
                gsai2['Sex']=gsai2['Sex'].replace({key:'M'})
            else: 
                gsai2['Sex']=gsai2['Sex'].replace({key:'-'})
cambioSexKey(keysSex)
gsai2['Sex'].value_counts()
```

###RESULTADOS###
